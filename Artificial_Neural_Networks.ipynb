##**<font color = green>4.Understanding how Artificial Neural Network works</font>**

Artificial Neural Networks constitutes of primarily 3 layers:

1.   Input layer
2.   Hidden Layer/Dense layer (can be multiple)
3.   Output layer

All the layers contain neurons which hold information about the data.
Unlike Machine Learning here the data is passed as an array to the input.
The data comprises of multiple columns combined into 1D array.

Weights are assigned to neurons, which indicates how much importance a neuron gives to a particular attribute (column).

Unlike Machine Learning models, Artificial neural networks train & test the data simultaneously. In addition to that, the data is fed row wise instead of an entire chunk of split train data.

This means, the first entry to the the system is the first row of the 1D array. Initially, random weights will be assigned to the neurons and these neurons will then transfer the information from input layer to the hidden layer and predict the output for the output layer. This predicted output will be then compared with the actual output, which will generate an error. This error is called as a 'Loss parameter'. The Loss parameter will be then fed back to the input along with the next row-wise entry. This time the neurons will tune their weights giving and include bias as well and the machine will be again trained on this data. Once again a loss parameter will be generated and fed back to the system along with the next input. Neurons will tune again & bias will be included. This will continue till all the entries of the dataset are trained. **On completion of one cycle on the entire dataset we say the machine has completed one epoch**. More number of epochs will make the machine much robust, since it will be well adapted to learn from it's previous mistakes.

However, while doing so one must remember that too many epochs may also lead to overfitting and too less may lead to underfitting. For those who don't know, Overfitting is a state where a machine trains so much on the training data that it predicts poorly on testing data especially when we have a high variance, while Underfitting is a state where machine trains poorly on training data giving inadequate performance both on training and testing data.



##**<font color = green>5.What is Bias & Activation function?</font>**

Now there is a thin line between weights and bias. Weights tell us the individual importance of a given attribute whereas bias is the tendency of a model to give importance to one of the attributes based on it's learning.
Bias is used for shifting the activation function to left or right and is called as the y-intercept in the line equation. To explain it simply, the artificial neural networks takes input and compute the weighted sum of the inputs and includes the bias as well. It is given in the form of a transfer function. An **activation function** on the other hand **decides whether the neuron should be activated or not**. This will decide whether the neuron's input is important or not in predicting the output. So in a way only selected neurons who contribute to good prediction are activated in each training step, or in simple words Activation functions incorporates bias in the system.

There are different types of activation functions like Linear activation function, SoftMax, Relu, Sigmoid, tanh.

Linear activation function is used for Regression.

Sigmoid is used for binary classification.

Softmax is used for multi-class classification.

Relu is used for Image classification (CNN).

Tanh is used for text classification (RNN).


In absence of an Activation function, neurons perform linear transformations on the input, **Activation function adds non-linearity and bias to the entire model**.

##**<font color = green>6.What is Learning rate?</font>**

Learning rate is how fast an algorithm gets trained. It depends on how fast the information is transferred between neurons. A learning rate low indicates that the machine will take time to learn. Higher learning rate requires high computational GPU and hence a high budget.

##**<font color = green>7. Feedforward vs Feedback neural networks</font>**

In Feed forward neural networks only the error is passed to the next input and not the output of the first prediction and in Feed back neural networks along with the error, the output of the first prediction is also fed back to the system.

Example of Feedforward system is CNN.
Example of Feedback system is RNN.

##**<font color = green>8. What is empirical loss?</font>**

Empirical loss is basically  the combined loss of of an entire epoch. Losses in neural networks are not calculated as row-wise but we get to know the combined loss of an entire epoch.

Here the output f(x) is the sum of weighted averages (x1W1 + x2W2) which is passed as an output along with a bias 'b'.

          So output, f(x) = ∑(x1w1 + x2W2) + b

Coming to the diagram now, here f(x) is the predicted o/p & y is the actual output. The difference between both is loss parameter.

          Loss parameter = y - f(x)

Empirical loss is the summation of all the Loss parameters for 1 epoch.

For a n-dimensional array,

          Empirical loss = ∑(Loss paramteres from 1 to nth row)

##**<font color = red>2. Importing the libraries & reading the data</font>**





We import the essential libraries form Keras. Sequential & Dense are the ones that will help us further in creating the model.

```python
from keras.models import Sequential
from keras.layers import Dense
import numpy
```

```python
# fix random seed for reproducibility
seed = 7
numpy.random.seed(seed)
# Load the  pima indians dataset
dataset = numpy.loadtxt("https://raw.githubusercontent.com/slmsshk/pima-indians-diabetes.data.csv/main/pima-indians-diabetes.csv", delimiter=",")

X= dataset[:,0:8]
Y = dataset[:,8]
```

In the code above we are using

**numpy.random.seed(seed)**: This line sets the seed for NumPy's random number generator. By setting the seed to the value assigned to seed (which is 7 in this case), the sequence of random numbers generated by NumPy's random functions will be the same each time you run the code. This ensures reproducibility in random number generation using NumPy.

**dataset = numpy.loadtxt("https://raw.githubusercontent.com/slmsshk/pima-indians-diabetes.data.csv/main/pima-indians-diabetes.csv", delimiter=","):** This line loads a dataset into a NumPy array named dataset. The dataset is fetched from a URL provided within the loadtxt function. This dataset appears to be related to diabetes among Pima Indians. The delimiter="," argument specifies that the data is separated by commas.

**X = dataset[:,0:8]:** This line extracts a subset of columns from the dataset array and assigns it to X. Specifically, it selects all rows (: means all rows) and the first 8 columns (0:7).

**Y = dataset[:,8]:** This line extracts a single column (column index 8) from the dataset array and assigns it to Y. This typically represents the target variable or labels associated with the observations in X. In this case, it seems to be the first column of the dataset, which represent whether or not a person has diabetes.

##**<font color = red>3. Creating the model</font>**

After reading the dataset, it is time to create the model.
Creating a model is done by defining the input layer, dense layer (hidden layer) and output layer.

```python
# create model
model = Sequential()
model.add(Dense(20,input_dim=8,activation = 'relu')) # first layer (i/p layer)
model.add(Dense(10,activation = 'relu')) # second layer
model.add(Dense(10,activation = 'relu')) # third layer
model.add(Dense(1,activation = 'sigmoid')) # fourth layer (o/p layer)
```

In the code above we are creating a Sequential model.
Firstly, we define an input layer with dimensions/attributes/columns = 8 and use activation function as relu.
In the input layer, the dense layer will come by default. Initially I have defined 20 dense layers.
It is then followed by adding 10 + 10 i.e. 20 more hidden (dense layers) to the model with the same activation function.
Lastly, we add just 1 output layer with activation function as 'sigmoid'.

One thing to note here is that we have used sigmoid function as activation for the output which indicates the output is binary classified indicating just 2 outputs.

##**<font color = red>4. Compiling the model</font>**

Once we design our model, the very next step is to compile it. The compilation can be done considering 3 parameters namely **loss, optimizer & metrics**.

```python
#Compiling model
model.compile(loss='binary_crossentropy',optimizer = 'adam',metrics = ['accuracy'])
```

**model.compile**: This is a method used in TensorFlow or Keras to compile a model. Compiling a model means configuring it for training. During compilation, you specify additional settings required to train the model, such as the optimizer, loss function, and evaluation metrics.

**loss='binary_crossentropy':** This specifies the loss function to be used during training. In this case, it's set to 'binary_crossentropy'. Cross-entropy is a commonly used loss function in binary classification problems, where the output of the model is binary (e.g., yes/no, 0/1). It measures the difference between the true labels and the predicted probabilities.

**optimizer='adam':** This specifies the optimizer to be used during training. The optimizer is responsible for updating the weights of the neural network during training in order to minimize the loss function. 'Adam' is a popular optimization algorithm in deep learning that combines ideas from RMSProp and Momentum.

**metrics=['accuracy']:** This specifies the metric(s) to be used to evaluate the performance of the model during training and testing. In this case, it's set to 'accuracy', which is a common metric for classification problems. Accuracy measures the proportion of correct predictions out of all predictions made by the model.

Important things to remember here is that if data is categorical the loss would be categorical_binary_crossentropy or for non categorical classification dataset we use binary_crossentropy.

The optimizer used here is 'adam'.

The metrics used here is accuracy. Some other metrics can be loss, validation loss or validation accuracy.

To check information about the model we use '.summary()' method.

```python
model.summary()
```

##**<font color = red>6. Fitting the model without Early stopping</font>**

Similar to a machine Learning model, in Deep Learning as well we fit the model.

```python
history = model.fit(X,Y, validation_split = 0.33, epochs = 250, batch_size = 10)
```

Now let us understand the output above, as mentioned earlier the output shows epoch wise:**loss, accuracy, validation & validation accuracy which are the 4 important metrics**.

If you look closely at the output above we have got a very good accuracy at the last epoch around 68 %, however if you closely look at epoch 52 we did obtain the same accuracy. So, it would have been better to stop the model at epoch 52 instead of training it for 250 epochs.

This concept about regularization that stops the training process of a neural network before it reaches maximum number of epochs is called as '**early stopping**'. The idea is to monitor the performance on the validation set and stop the training when the validation error starts to increase.

More epoch cycles as mentioned earlier may cost high computational GPU and a high expense and hence early stopping is important for regularisation.

##**<font color = red>7. Evaluating the model</font>**

So after training the model we calculate the loss & accuracy.

```python
# evaluate the model
scores = model.evaluate(X,Y)
print("%s: %.2f%%" % (model.metrics_names[1],scores[1]*100))
```

In the above code we are evaluating the model on X & Y.

**print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100)):** This line prints out one of the performance metrics of the model. It uses string formatting (% operator) to format the output.

**model.metrics_names[1]:** This retrieves the name of the metric being printed. It accesses the second element of the model.metrics_names list (i.e. accuracy) assuming that the first element represents a different metric (like loss).

**scores[1]*100:** This retrieves the value of the corresponding metric from the scores variable and multiplies it by 100 to convert it to a percentage. The index [1] is used assuming that the first metric is something like loss, and the second metric is the one being printed (e.g. accuracy).

**%.2f%%:** This is a format specifier for a floating-point number with two decimal places (%.2f) followed by a percentage sign (%%).

```python
# Visualize training history
# list all the data in history
history.history.keys()
```

history.history.keys() returns a list of all the metric names that were monitored during the training of the model. This information can be useful for analyzing and visualizing the performance of the model during training.

##**<font color = red>6. Fitting the model with Early stopping</font>**

 Let us use a different approach now. **This time we will incorporate 'Early stopping'**.

 If one trains the model directly it may take more epochs (that's not a problem) but there might be a scenario where the accuracy, loss, validation accuracy & validation loss are stable after a few epochs and as the epochs proceed later these metrics show a  lesser value.

In such scenarios we can train the model to do an early stopping so that we get optimum  value of these metrics [accuracy, loss, validation accuracy & validation loss].

Let us import EarlyStopping from keras.callbacks library first.

```python
from keras.callbacks import EarlyStopping
```

We have to specify what exactly we are monitoring, in the code below I am monitoring 'val_loss'. This indicates as soon as my val_loss becomes stable, the model will stop it's training.

```python
early_stopping = EarlyStopping(monitor='val_loss', patience=10, verbose=1, restore_best_weights=True)
```

optimum metrics for our model.

```python
history = model.fit(X, Y, validation_split=0.33, epochs=250, batch_size=10, callbacks=[early_stopping])
```

Here instead of performing 250 epochs our machine has stopped at epoch 27. Thus avoiding the model to train till 250 epochs.
